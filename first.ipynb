```python
import os

from groq import Groq

client = Groq(
    api_key=os.environ.get("GROQ_API_KEY"),
)
```

```python
messages = [{
    "role":"system",
    "content":"what ever user ask you should answer right"
},{
    "role":"user",
    "content":"what is the capital of india?"
},
{
    "role":"user",
    "content":"what is the capital of japan?"
}
]
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=messages
)
print(response.choices[0].message.content);
```

```python
system_prompt = "Your are helpful assistant whatever scraped data user give of the website summerize that content in 100 words and it should be understanable and clear.Answer should be in mutiple line so that it stays within page viewpoint"
```

```python
user_promt = "Please give the details of the website in readable manner."
```

```python
def get_website_data(website):
    return[
        {
            "role":"system",
            "content":system_prompt
        },
        {
            "role":"user",
            "content":user_promt + website
        },
    ]
```

```python
from fetchContent import fetch_website_contents

website_content = fetch_website_contents("https://www.altnews.in/cockroach-janata-partys-majority-followers-pakistani-bjym-secretary-tajinder-bagga-others-amplify-misleading-claim/")

# print("website_content",website_content)

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=get_website_data(website_content)
)
print(response.choices[0].message.content)
```

```python
email_system_prompt = """
You are a high-accuracy email summarization assistant.

Your task is to summarize email content into concise and well-structured bullet points.

Strict Rules:
- Ignore the email subject completely.
- Do NOT include the subject line in the response.
- Do NOT mention or summarize the subject.
- Start directly with bullet points from the email body content only.
- Output strictly in bullet point format.

Instructions:
- Extract all important information from the email body.
- Include deadlines, action items, decisions, responsibilities, risks, and important context.
- Keep the summary concise but information-dense.
- Avoid repetition, greetings, signatures, and unnecessary filler content.
- Preserve the original meaning and intent.
- Do not hallucinate or invent information.
- Use clear and professional language.
"""
```

```python
user_system_prompt = "Please summerize given email for me"
```

```python
def get_email_content(content):
    return [{
        "role":"system",
        "content":"email_system_prompt"
    },
    {
        "role":"user",
        "content":"user_system_prompt" + content
    },
    ]
```

```python
from fetchContent import email_content

body = "\n".join(email_content.split("\n")[2:])
# print(body)

response =  client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=get_email_content(body)
)
print(response.choices[0].message.content)
```
